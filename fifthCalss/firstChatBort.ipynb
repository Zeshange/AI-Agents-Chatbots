```python
# Import libraries
import google.generativeai as genai
import pandas as pd
import PyPDF2

# Configure Gemini API
genai.configure(api_key="AIzaSyCM4vIdZylsML_EvYub0ky-ynPuJtYXvUE")

# --- Load Knowledge Base ---
knowledge_text = ""

# Read PDF content (if available)
try:
    with open("company_policies.pdf", "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        for page in pdf_reader.pages:
            knowledge_text += page.extract_text() + "\n"
except FileNotFoundError:
    print("No PDF found. Skipping PDF import.")

# Read CSV content (if available)
# try:
#     csv_data = pd.read_csv("employee_data.csv")
#     knowledge_text += "\n" + csv_data.to_string()
# except FileNotFoundError:
#     print("No CSV found. Skipping CSV import.")

# --- Chatbot Function ---
def knowledge_bot(user_query):
    prompt = f"""
You are a knowledge-based assistant. Answer based on the following data:
{knowledge_text}

User: {user_query}
Assistant:
"""
    model = genai.GenerativeModel("gemini-2.0-flash")
    response = model.generate_content(prompt)
    return response.text

# --- Run Chatbot ---
print("Knowledge Bot Ready! Type 'exit' to quit.")
while True:
    question = input("You: ")
    if question.lower() in ["exit", "quit"]:
        print("Assistant: Goodbye!")
        break
    answer = knowledge_bot(question)
    print(f"Assistant: {answer}")
```

Travel Bot

What is it?
A smart travel assistant chatbot that helps users find flights, hotels, and the best travel time based on a CSV database.

What does it do?
It provides personalized travel suggestions from travel_data.csv for any country or city the user asks about.

Key Features:

Suggests hotels with prices

Recommends flights with fares

Advises on the best travel seasons

How does it work?
It reads travel information from a CSV file, processes your query using Gemini AI, and gives relevant travel advice.

How to use?
Run the script, ask travel-related questions (e.g., â€œBest time to visit Dubai?â€, â€œHotels in Paris?â€), and type exit to quit.

```python

import google.generativeai as genai
import pandas as pd
# Configure Gemini API
genai.configure(api_key="AIzaSyCM4vIdZylsML_EvYub0ky-ynPuJtYXvUE")

# Load Travel Data
try:
    travel_df = pd.read_csv("travel_data.csv")
    travel_data = travel_df.to_string(index=False)  # Convert CSV into text format for AI
except FileNotFoundError:
    travel_data = "No travel data available."
    print("âš ï¸ travel_data.csv not found. Using default prompt.")

# --- Travel Chatbot Function ---
def travel_chatbot(user_query):
    prompt = f"""
You are a travel assistant.
Your task: Only suggest options from the following database:
{travel_data}

User asked: {user_query}
Provide only relevant flights, hotels, and best travel time.
"""
    model = genai.GenerativeModel("gemini-2.0-flash")
    response = model.generate_content(prompt)
    return response.text

# --- Run the Travel Chatbot ---
print("ðŸŒ Travel Chatbot Ready! Type 'exit' to quit.")
while True:
    question = input("You: ")
    if question.lower() in ["exit", "quit"]:
        print("Chatbot: Safe travels! Goodbye!")
        break
    answer = travel_chatbot(question)
    print(f"Chatbot: {answer}")
```
